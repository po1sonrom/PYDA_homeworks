***Решение***

```python
power = pd.read_csv('power.csv')

baltic_countryies = power[(power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] == 'Estonia') ] #берем информацию по Латвии, Литве и Эстонии
fc = baltic_countryies[(baltic_countryies['category'] == 4) | (baltic_countryies['category'] == 12) | (baltic_countryies['category'] == 21) & ( baltic_countryies['year'] < 2010) & (baltic_countryies['year'] > 2005)] #делаем нужную выборку по категориям и годам
fc['quantity'].sum() #считаем подтребление сумарное по столбцу quantity
```
