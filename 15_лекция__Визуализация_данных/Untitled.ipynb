```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = pd.read_csv(
    'googleplaystore.csv'
)

names.shape
```

```python
names = names.dropna() #убрали пустые строки. Их было не очень много
names.shape
```

```python
names.head()
```

```python
aa = names.Rating.hist(bins=10)
aa.set_title('Оценка рейтинга')
```

```python
total_category = names['Category'].value_counts()
total_category
#total_category.hist()
```

```python
free_paid = names['Type'].value_counts()
    
free_paid.plot(kind='pie', y='Count')
```

```python
names['Reviews'] = names.Reviews.astype('int64')
```

```python
top_developers_df = names.groupby('Category')[['Reviews']].sum()\
    .sort_values('Reviews', ascending=False)
    
top_developers_df
```

```python
top_developers_df.style.bar()
```
