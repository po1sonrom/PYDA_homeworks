### 5. С использованием t-test статистики проверить взаимосвязи между несколькими выборками (двумя или больше)

```python
print(f'Взаимосвязь между среднемесячными отработанными часами и последней оценкой: \nlast_evaluation: {sts.shapiro(hr["last_evaluation"])} \naverage_montly_hours: {sts.shapiro(hr["average_montly_hours"])}')
```

*Необходимо взять больше выборки, так как идет стремление к 1, но не явно выраженное.  
Проверка на равность дисперсии:*

```python
statistics.variance(hr['last_evaluation']), statistics.variance(hr['average_montly_hours'])
```

*В связи с отсутствием равенства дисперсии, делаем проверку на t-тест:*

```python
sts.ttest_ind(hr['last_evaluation'], hr['average_montly_hours'], equal_var=False )
```

*Из-за того, что p-value равно 0, а t-value имеет очень большое отрицательно значение, то получается, что данные переменных не взаимосвязаны.*

### 10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

#### - Первая гипотеза

Причина увольнения не зависит от получаемой зарплаты

```python
print(stats.shapiro(hr.satisfaction_level))
```

*Проверяем размерность на выборках*

```python
left_true = hr[hr['left'] == 1]['satisfaction_level']
left_false = hr[hr['left'] == 0]['satisfaction_level']
print(f'Уволенные сотрудники: {len(hr[hr["left"] == 1])}')
print(f'Работающие сотрудники: {len(hr[hr["left"] == 0])}')
```

*Уравниваем длины*

```python
left_true = left_true.sample(3500)
left_false = left_false.sample(3500)
```

*Дисперсия*

```python
statistics.variance(left_true), statistics.variance(left_false)
```

*Применяем t-test*

```python
print(stats.ttest_ind(left_true, left_false, equal_var=False))
```

*Из-за отрицательной корреляции между переменными приходим к выводу, что уровень удовлетворенности уволенных ниже, чем работающих.*

#### - Вторая гипотеза  
У ушедших сотрудников такая же оценка, что и оставшихся.

*Проверка распределения переменных.*

```python
sts.shapiro(hr['last_evaluation'])
```

*Создаем выборки*

```python
left_ls = hr[hr['left']== 1]['last_evaluation']
left_no_ls = hr[hr['left']== 0]['last_evaluation']

len(left_ls), len(left_no_ls)
```

*Так как длины разные, их необходимо уровнять*

```python
left_ls = left_ls.sample(3500)
left_no_ls = left_no_ls.sample(3500)
```

*Дисперсия*

```python
statistics.variance(left_ls), statistics.variance(left_no_ls)
```

*Применяем t-тест*

```python
sts.ttest_ind(left_ls,left_no_ls, equal_var=False )
```

*Строим график для наглядности*

```python
sns.distplot(left_ls, 20, color="y")
sns.distplot(left_no_ls, 20, color="g")
plt.title('Плотность вероятности распределения последней оценки')
plt.legend(['left_ls', 'left_n0_ls'])
plt.show()
```

*В связи с тем, что t-value меньше 1, а pvalue значительно больше 0,05. Различий у группы покинувших компанию и оставшихся работать нет значимых различий в оценке.*
